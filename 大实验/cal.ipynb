```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
for o in range(2,7):
    f=open("5.20-"+str(o)+".in","r")
    n=1800
    t=[0]*n
    a=[0]*n
    b=[0]*n
    dx=[[] for _ in range(n)]
    x=[0]*n
    x2=[0]*n
    SE=[0]*n
    for i in range(n):
        s=f.readline().split()
        # print(s)
        a[i]=float(s[1])*1000
        b[i]=float(s[2])
        t[i]=float(s[0])

    for i in range(1,60):
        for j in range(i,n,i):
            dx[i].append((a[j]-a[j-i]))
    df=pd.DataFrame(dx[0:9])



    for i in range(n):
        x[i]=np.mean(dx[i])
        x2[i]=np.mean([i*i for i in dx[i]])
        SE[i]=(np.var(dx[i]))**0.5
    df=df.T
    df=df.join(pd.DataFrame({"<x>":x[1:9]}))
    df=df.join(pd.DataFrame({"<x^2>":x2[1:9]}))
    df=df.join(pd.DataFrame({"SE":SE[1:9]}))
    df.to_excel("5.20-"+str(o)+"(cal,1800).xlsx",index=False)
    beta,apha= np.polyfit(t[1:60],x2[1:60], 1,cov=False)
    print(format(apha,'.4f'),'+',format(beta,'.4f'),'\\lambda_{Hg}^0')
    plt.plot(range(1,60),x[1:60],label="<X>")
    plt.plot(range(1,60),x2[1:60],label="<X^2>")
    plt.plot(range(1,60),[i/10*beta+apha+2*SE[i] for i in range(1,60)],label="upper")
    plt.plot(range(1,60),[i/10*beta+apha for i in range(1,60)],label="mid")
    plt.plot(range(1,60),[i/10*beta+apha-2*SE[i] for i in range(1,60)],label="down")
    # plt.plot(t,x3,label="var")
    # plt.plot(a,b,label="a,b")
    plt.xlabel("time")
    plt.ylabel("value")
    plt.title("mean and var")
    plt.legend()
    plt.show()
    f.close()
```
